```python
import gymnasium as gym
from text_flappy_bird_gym import *
# from flappy_bird_gymnasium import *

from collections import defaultdict
import csv
import numpy as np
import os, sys
import time
```

```python
env = gym.make('TextFlappyBird-v0', height=20, width=30, pipe_gap=2, seed=None)
env.reset()
```

```python
# Get value from .csv file to dict Q
def convert_Q_(namefile):
    Q = defaultdict(lambda: 0)
    
    with open(namefile, mode ='r')as file:
        csvFile = csv.reader(file)
        for line in csvFile:
            [a,b,c,d] = line
            a,b,c = int(float(a)), int(float(b)), int(float(c))
            Q[((a,b), c)] = float(d)
    return Q
```

```python
# Edit name file
Q = convert_Q_('Q_Monte_Carlo.csv')
```

```python
def run(env):
    state,_ = env.reset()
    done = False
    episode_reward = 0
    while not done:
        action = np.argmax([Q[state,0],Q[state,1]])
        next_state, reward,done,_, info = env.step(action)
        episode_reward += reward
        state = next_state
        if info['score'] >1e2: #Stopping condition to avoid it to be stuck if it never dies
            break
        os.system("clear")
        sys.stdout.write(env.render())
        time.sleep(0.2) # FPS
    return info['score']
```

```python
env = gym.make('TextFlappyBird-v0', height=20, width=30, pipe_gap=2, seed=None)
env.reset()
run(env)
```
